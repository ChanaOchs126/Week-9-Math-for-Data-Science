```python
import pandas as pd

# Load both datasets
df_qb = pd.read_csv('qb_stats.csv')
df_survey = pd.read_csv('survey_data.csv')

print("DataFrames df_qb and df_survey loaded successfully.")
print("\nFirst 3 rows of QB Data:")
print(df_qb.head(3).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Part 1.a: Calculate Mean
numerical_cols = ['games', 'att', 'cmp', 'comp_pct', 'yds', 'ypa', 'td', 'int', 'sack', 'loss', 'rating']
mean_values = df_qb[numerical_cols].mean()

print("**(a) Mean of each numerical column:**")
print(mean_values.to_markdown(numalign="left", stralign="left"))
```

```python
# Part 1.b: Calculate Standard Deviation
std_values = df_qb[numerical_cols].std()

print("**(b) Standard deviation of each numerical column:**")
print(std_values.to_markdown(numalign="left", stralign="left"))
```

```python
# Part 1.c: Histogram of Yards (Textual Output)
print("**(c) Histogram of Yards Data (Textual Output):**")
print("Frequency distribution of yards for plot:")
print(df_qb['yds'].value_counts(bins=5, sort=False).to_markdown(numalign="left", stralign="left"))
```

```python
# Part 1.d: Boxplot of Completion Percentage Data (Textual Output)
print("**(d) Boxplot of Completion Percentage Data (Textual Output):**")
print("Five-Number Summary for Comp. Pct. boxplot:")
print(df_qb['comp_pct'].describe().loc[['min', '25%', '50%', '75%', 'max']].to_markdown(numalign="left", stralign="left"))
```

```python
# Setup for Probability Calculations
hair_eye_table = pd.crosstab(df_survey['hair_color'], df_survey['eye_color'])
total_students = df_survey.shape[0]

brown_hair_count = hair_eye_table.loc['brown'].sum()
blue_eyes_count = hair_eye_table['blue'].sum()
blue_eyes_and_brown_hair_count = hair_eye_table.loc['brown', 'blue']

print(f"Total Students (N): {total_students}")
print(f"Count(Brown Hair): {brown_hair_count}")
print(f"Count(Blue Eyes): {blue_eyes_count}")
```

```python
# Part 2.a: P(Brown Hair)
p_brown_hair = brown_hair_count / total_students
print(f"**(a) P(Brown Hair): {p_brown_hair:.4f} ({brown_hair_count}/{total_students})**")

# Part 2.b: P(Blue Eyes)
p_blue_eyes = blue_eyes_count / total_students
print(f"**(b) P(Blue Eyes): {p_blue_eyes:.4f} ({blue_eyes_count}/{total_students})**")
```

```python
# Part 2.c: P(Blue Eyes | Brown Hair)
p_blue_eyes_given_brown_hair = blue_eyes_and_brown_hair_count / brown_hair_count
print(f"**(c) P(Blue Eyes | Brown Hair): {p_blue_eyes_given_brown_hair:.4f} ({blue_eyes_and_brown_hair_count}/{brown_hair_count})**")
```

```python
# Part 2.d: P(Brown Hair | Blue Eyes)
p_brown_hair_given_blue_eyes = blue_eyes_and_brown_hair_count / blue_eyes_count
print(f"**(d) P(Brown Hair | Blue Eyes): {p_brown_hair_given_blue_eyes:.4f} ({blue_eyes_and_brown_hair_count}/{blue_eyes_count})**")
```

```python
import numpy as np

# Part 2.e: Check for Independence
is_independent_check = np.isclose(p_blue_eyes_given_brown_hair, p_blue_eyes)

print("**(e) Independence Check: **")
print(f"   P(Blue Eyes | Brown Hair) = {p_blue_eyes_given_brown_hair:.4f}")
print(f"   P(Blue Eyes)            = {p_blue_eyes:.4f}")

if is_independent_check:
    print("Result: Events ARE independent.")
else:
    print("Result: Events ARE NOT independent.")
    print("Conclusion: The events are dependent because P(Blue Eyes | Brown Hair) is NOT equal to P(Blue Eyes).")
```
